```python
# Приведем названия полей к общепринетому виду
# Создадим функцию, где заменим пробелы на нижнее подчеркивание и приведем к нижнему регистрцу
def replace_lowercase(column):
    return column.replace(' ', '_').lower()
bookings = bookings.rename(columns = replace_lowercase)
```

```python
bookings.head()
```

```python
# Отберем только те записи, которые не отменены
# Сгрупируем по странам
# Посчитаем количесвто бронирований
# Отсроритруем по убыванию
# Выведем первые 5 записей данного запроса

bookings \
    .query('reservation_status_date != "Canceled"') \
    .groupby('country') \
    .agg({'hotel':'count'}) \
    .sort_values('hotel', ascending=False) \
    .head()                                           
```

```python
# Узнаем среднее количесвто дней, проведенных в отелях
bookings.groupby('hotel').agg({'stays_total_nights':'mean'}).round(2)
```

```python
# Узнаем через query: сколько было несоответствий присвоенных и забронированных номеров 
bookings.query('reserved_room_type != assigned_room_type').agg({'hotel': 'count'})
```

```python
bookings.head()
```

```python
# Выясним в какой месяц чаще всего бронировали отель в 2016 году
bookings \
    .query('arrival_date_year == 2016') \
    .groupby('arrival_date_month', as_index=False) \
    .agg({'hotel': 'count'}) \
    .sort_values('hotel', ascending=False)
```

```python
# Выясним в какой месяц чаще всего бронировали отель в 2017 году
bookings \
    .query('arrival_date_year == 2017') \
    .groupby('arrival_date_month', as_index=False) \
    .agg({'hotel': 'count'}) \
    .sort_values('hotel', ascending=False)
```

```python
# Количество отмененных бронирований по годам и месецам
bookings \
    .query('is_canceled == 1') \
    .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
    .agg({'hotel':'count'}) \
    .sort_values('hotel', ascending=False) \
    .head(6)
```

```python
# Выведем среднее значение по данным колонкам
bookings[['adults', 'children', 'babies']].mean()
```

```python
# Создадим колонку total_kids, где объединим колонки children и babies
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
# Выведем среднее количество детей по отелям
bookings.groupby('hotel').agg({'total_kids': 'mean'}).round(2)
```

```python
# Создадим колонку, которое принимает True, если есть дети в бронировании, или False, если нет детей в бронировании
bookings['has_kids'] = bookings['total_kids'] > 0
```

```python
# Количество отмененных бронирований без детей
bookings.query('is_canceled == 1 and has_kids == False').shape[0]
```

```python
# Общее количество бронирований без детей
bookings.query('has_kids == False').shape[0]
```

```python
# Отношение между отмененными бронированиями с детьми к общшему количечсству бронирований с детьми
bookings.query('is_canceled == 1 and has_kids == False').shape[0] / bookings.query('has_kids == False').shape[0]
```

```python
# Укажем получившееся значение выше в процентах
churn_rate = bookings.query('is_canceled == 1 and has_kids == False').shape[0] / bookings.query('has_kids == False').shape[0]
round(churn_rate * 100, 2)
```
